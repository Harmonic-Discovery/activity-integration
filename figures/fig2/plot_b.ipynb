```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 

data = pd.read_csv("../../data/prepped_activities_all_chembl_pubchem.csv")
measured = data[data["activity_type"] == "measured"]["activity_value"].tolist()
inferred = data[data["activity_type"] == "inferred"]["activity_value"].tolist()

sns.set_style("white")

plt.figure(figsize=(5.,3.5))
sns.kdeplot(inferred, fill=True, color="#337CA7", label="Inferred Activities From POC")
sns.kdeplot(measured, fill=True, color="#E96274", label="Measured Activities\n(ChEMBL + PubChem)")
plt.vlines(6.0, 0., 10.0, color="black", linestyles="--")
plt.legend(fontsize=9.5, loc="upper right")
plt.ylim(0, 3.8)
plt.xlim(3.5, 10)
plt.text(4, 1, "Inactive", fontdict={"color": "red"})
plt.text(7, 1, "Active", fontdict={"color": "green"})
# plt.title("Distribution of Inferred and Predicted Activites", fontsize=14)
plt.xlabel(r"pIC$_{50}$ Value", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.tight_layout()
plt.savefig("measured_vs_inferred_activity.svg", format="svg")
plt.show()
```
