```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt 

plt.figure(figsize=(6,4))
data = pd.read_csv("predicted_activity_100nM_1000nM.csv")
data["100nM"] = data["100nM"].clip(0, 100)
data["1000nM"] = data["1000nM"].clip(0, 100)
plt.scatter(x=data["100nM"], y=data["pred"], label="@ 100nM", color="#25719E", alpha=0.3, s=3)
plt.scatter(x=data["1000nM"], y=data["pred"], label="@ 1000nM", color="#D13F58", alpha=0.3, s=3)
plt.xlabel("% Inhibition", fontsize=14)
plt.ylabel("Predicted pActivity", fontsize=14)
plt.legend(markerscale=2)
plt.tight_layout()
plt.savefig("pred_activity_vs_inhibition_100nM_1000nM_scatter.png", dpi=1000)
plt.show()
```
